# Load Llama3.1 model

this model is used for text-generation

```python
#model_name = "meta-llama/Meta-Llama-3-8B-Instruct"
model_name = "/kaggle/input/llama-3.1/transformers/8b-instruct/1"
tokenizer = transformers.AutoTokenizer.from_pretrained(model_name,token= secret_value_0, device_map = 'auto')
model = transformers.AutoModelForCausalLM.from_pretrained(model_name,token= secret_value_0, device_map = 'auto',torch_dtype=torch.float16)
```

```python
pipe = transformers.pipeline('text-generation', 
                             model=model,
                             tokenizer=tokenizer,
                            )
```

```python
messages =[ {
    "role":"user",
    "content":"que fait la karmine corp",
},]

output = pipe(messages,pad_token_id=pipe.tokenizer.eos_token_id,max_new_tokens=5000)

messages = output[0]["generated_text"]

assistant_response = output[0]["generated_text"][-1]["content"].strip()

print(assistant_response)
```

# Load gemma model

```python
#model_name = "google/gemma-2-2b-it"
model_name = "/kaggle/input/gemma-2/transformers/gemma-2-2b-it/1"

tokenizer = transformers.AutoTokenizer.from_pretrained(model_name,token= secret_value_0, device_map = 'auto')
model = transformers.AutoModelForCausalLM.from_pretrained(model_name,token= secret_value_0, device_map = 'auto')
```
